```python
import pandas as pd
```

```python
#Load data to create the dataframe
URL='https://cocl.us/Geospatial_data/Geospatial_Coordinates.csv'
df_geo=pd.read_csv(URL)
df_geo.head()
```

```python
#Rename the column name to make the record tally
df_geo = df_geo.rename(columns = {'Postal Code':'PostalCode'}) 
df_geo.head()
```

```python
#Load the data where we had saved during Question 1
df3=pd.read_csv('Capstone_Q1.csv')
df3.head()
```

```python
#Drop the repeated column "Unnamed:0"
df3.drop('Unnamed: 0', axis='columns', inplace=True)
df3.head()
```

```python
#Merge two dataframes base on 'PostalCode' column
df3 = pd.merge(df3, df_geo, on = 'PostalCode')
df3.head()
```

```python
#Save the data for later use in Question 3
df3.to_csv('Capstone_Q2.csv', index=False)
print('Successfully Saved!')
```
